```python
missing_values=["N/a", "na", np.nan]
```

```python
data = pd.read_csv("weatherHistory.csv", na_values=missing_values)
```

```python
df = data.drop(['Summary', 'Precip Type', 'Temperature (C)', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Visibility (km)', 'Daily Summary'], axis=1)
df.head()
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df = df.set_index('Formatted Date')
data = df[['Apparent Temperature (C)', 'Humidity', 'Pressure (millibars)']].resample('MS').mean()
data
```

```python
plt.figure(figsize=(10,10))
plt.xlabel('TEMPERATURE')
plt.ylabel('PRESSURE')
plt.title('GAY LUSSAC\'s LAW')
plt.barh(data['Apparent Temperature (C)'], data['Pressure (millibars)'], linewidth=1)
plt.show()
```

```python
#plotting whole dataset for all months
plt.figure(figsize=(15,3));
plt.plot(data['Humidity'], label = 'Humidity', color='orange', linestyle='dashed')
plt.plot(data['Apparent Temperature (C)'], label = 'Apparent temp.', color='blue')
plt.title('Variation of Apparent temperature v/s Humidity', fontsize=25);
plt.legend(loc=0, fontsize=12)
plt.xticks(fontsize=15)
plt.yticks(fontsize=13)
```

```python
#plotting for a specific month
april = data[data.index.month==4]
```

```python
plt.figure(figsize=(15,3));
plt.plot(april.loc['2006-04-01':'2016-04-01', 'Apparent Temperature (C)'], marker='o', linestyle='-', label='Month of April')
plt.plot(april.loc['2006-04-01':'2016-04-01', 'Humidity'], marker='o', linestyle='-', label='Humidity', color='pink')
plt.legend(loc=0, fontsize=15)
plt.xlabel('Month of April', fontsize=15)
plt.title('Humidity v/s Apparent Temperature', fontsize = 15)
```
